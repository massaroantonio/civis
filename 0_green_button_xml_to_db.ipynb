```python
#returns a dictionary with the parameters of the start and end rules for dst offset
def dst_offset_parameters(dst_rule):
	binary_dst_rule=bin(int(dst_rule,16))[2:].zfill(32)
	start=[0,4,7,12,15,20]
	end=[4,7,12,15,20,31]
	labels=['month','operator','day_of_the_month','day_of_the_week','hours','seconds']
	dst_rule_dictionary={labels[i]: int(binary_dst_rule[start[i]:end[i]],2) for i in range(len(start))}
	return dst_rule_dictionary
```

```python
#returns datetime at which dst switches
def dst_switch(dst_rule, year):
	dst_rule_dictionary=dst_offset_parameters(dst_rule)
	month=dst_rule_dictionary['month']
	operator=dst_rule_dictionary['operator']
	day_of_the_month=dst_rule_dictionary['day_of_the_month']
	day_of_the_week=dst_rule_dictionary['day_of_the_week']
	hours=dst_rule_dictionary['hours']
	seconds=dst_rule_dictionary['seconds']
	
	
	if operator==0:
		out=datetime.datetime(year,month,day_of_the_month,hours,int(seconds/60))

	elif operator==1:
		out=datetime.datetime(year,month,day_of_the_month,hours,int(seconds/60))
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
	
	elif operator==2:

		out=datetime.datetime(year,month,1,hours,int(seconds/60))
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)		
	
	elif operator==3:

		out=datetime.datetime(year,month,1,hours,int(seconds/60))
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
		out=out+datetime.timedelta(days=1)
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
		
	elif operator==4:
		out=datetime.datetime(year,month,1,hours,int(seconds/60))
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
		out=out+datetime.timedelta(days=1)
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
		out=out+datetime.timedelta(days=1)
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)

		
	elif operator==5:
		out=datetime.datetime(year,month,1,hours,int(seconds/60))
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
		out=out+datetime.timedelta(days=1)
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
		out=out+datetime.timedelta(days=1)
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
		out=out+datetime.timedelta(days=1)
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)		

	elif operator==6:
		out=datetime.datetime(year,month,1,hours,int(seconds/60))
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
		out=out+datetime.timedelta(days=1)
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
		out=out+datetime.timedelta(days=1)
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
		out=out+datetime.timedelta(days=1)
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
		out=out+datetime.timedelta(days=1)
		while out.weekday()+1!=day_of_the_week:
			out=out+datetime.timedelta(days=1)
		
	elif operator==7:
		out=datetime.datetime(year,month+1,1,hours,int(seconds/60))-datetime.timedelta(days=1)
		while out.weekday()+1!=day_of_the_week:
			out=out-datetime.timedelta(days=1)

	else:
		assert(0)
		
	return out
```

```python
#uses a shell script to convert a timestamp into a datetime in the correct timezone and with the correct dst offset. 
#This is a work-around, since in python managing datetimes keeping track of the timezones and dst is a mess
def to_datetime(timestamp, tz_offset, dst_offset):
	ST='@'+str(timestamp)
	ps=subprocess.Popen(('echo',ST), stdout=subprocess.PIPE)
	output=subprocess.check_output('../../../../home/amassaro/./x.sh',stdin=ps.stdout)
	output=bytes.decode(output)[:-1]
	date=datetime.datetime.strptime(output, '%d-%m-%Y %H:%M:%S')				
	return date
```

```python
#creates an element of the form {"time_seconds":int,"time_date":string,"value":int, "duration":int} and stores it in the collection "user" inside the database db_name. the name of the collection is stored in the first entry of the list "fields", the other entries carry the values of the element to be stored
def insert_one_consumer_data_in_db(fields, db_name):
	client=MongoClient()
	db=client[db_name]
	i=0
	user=fields[0]
	element={"time_stamp":fields[1],"date_time":fields[2],"weekday":fields[3], "tz_offset":fields[4],"duration":fields[5], "value":fields[6]}
	db[user].insert(element)
	i=i+1
	#print("inserted line "+str(i))
	return
```

```python
#stores the data contained in a green button xml file in a collection named after the xml file itself, inside the database db_name
#at the moment it is way too slow because it calls the shell script for every date-time conversion. I am working to convert the
#timestamps into datetimes in batch
def load_gb_file_in_db(gb_file_name,db_name,xmlschema):
	import datetime
	fields=["N/A" for i in range(7)]
	user_id="user_"+gb_file_name.split("/")[-1][:-4]
	fields[0]=user_id
	xmlschema_doc=etree.parse(xmlschema)
	xmlschema=etree.XMLSchema(xmlschema_doc)

	tree=etree.parse(gb_file_name)
	root=tree.getroot()

	interval_block_list=[]
	namespace="http://naesb.org/espi"
	tz_offset=root.find(".//{http://naesb.org/espi}tzOffset").text	
	if (isinstance(tz_offset,str) and tz_offset!=""):
		tz_offset=int(tz_offset)
	else:
		tz_offset="N/A"

	dst_offset=root.find(".//{http://naesb.org/espi}dstOffset").text	
	if (isinstance(dst_offset,str) and dst_offset!=""):
		dst_offset=int(dst_offset)
	else:
		dst_offset="N/A"

	dst_start_rule=root.find(".//{http://naesb.org/espi}dstStartRule").text	
	if (isinstance(dst_start_rule,str) and dst_start_rule!=""):
		dst_start_rule=str(dst_start_rule)
	else:
		dst_start_rule="N/A"

	dst_end_rule=root.find(".//{http://naesb.org/espi}dstEndRule").text	
	if (isinstance(dst_end_rule,str) and dst_end_rule!=""):
		dst_end_rule=str(dst_end_rule)
	else:
		dst_end_rule="N/A"

	


	interval_block_list= root.findall(".//{http://naesb.org/espi}IntervalBlock")
	starts=[]
	ends=[]
	for block in interval_block_list:
		interval=block.find(etree.QName(namespace,"interval"))
		start=int(interval.find(etree.QName(namespace,"start")).text)
		duration=int(interval.find(etree.QName(namespace,"duration")).text)
		starts.append(start)
		ends.append(start+duration)
	
	for i in range(len(starts)):
		for j in range(i):
			assert(end[i]<start[j] or end[j]<start[i]), 'overlapping time periods for different IntervalBlocks'

	
	for block in interval_block_list:
		assert xmlschema(block), 'an interval-block does not comply with the xml schema for greenbutton xml files'
		print('ok interval block compatible')
		intervals=block.findall(".//{http://naesb.org/espi}IntervalReading")
		for interval in intervals:
			for x in interval.iter():
				if isinstance(x.tag, str):
					tag=etree.QName(x)

					if tag.localname=="start":
						if (isinstance(x.text, str) and x.text!=''):
							fields[1]=int(x.text)
							fields[2]=to_datetime(fields[1],tz_offset,0)#rifarlo in modo da fare un'unica chiamata, trasformare tutti i timestamp in datetime in una volta
							dst_start=dst_switch(dst_start_rule, fields[2].year)
							dst_end=dst_switch(dst_end_rule, fields[2].year)
							if (fields[2]>=dst_start and fields[2]<=dst_end):
								fields[2]=to_datetime(fields[1],tz_offset,dst_offset)
							fields[3]=fields[2].weekday()
						else:
							fields[1]='N/A'
							fields[2]='N/A'
							fields[3]='N/A'
					fields[4]=tz_offset
					if tag.localname=="duration":
						if (isinstance(x.text, str) and x.text!=''):
							fields[5]=int(x.text)  
						else:
						 	fields[5]='N/A'
					if tag.localname=="value":
						if (isinstance(x.text, str) and x.text!=''):
							fields[6]=int(x.text)
						else:
							fields[6]='N/A' 
			insert_one_consumer_data_in_db(fields, db_name)
			fields=["N/A" for i in fields]
			fields[0]=user_id

	interval_length=root.find(".//{http://naesb.org/espi}intervalLength").text
	if (isinstance(interval_length,str) and interval_length!=""):
		interval_length=int(interval_length)  
	else:
		interval_length="N/A"
	client=MongoClient()
	client[db_name][user_id].insert({"interval_length":interval_length})
	return
```

```python
#stores all the green button xml files inside a directory in the database db, each file will be stored in a different collection, named after the file itself. 
#reminder: try to store all the names inside load_gb_file_in_db() by appending to a list stored in a collection, it is more elegant
def load_all_gb_files_in_db(directory, db_name, xmlschema):   
	from os import listdir
	from os.path import isfile, join
	user_ids=[]
	onlyfiles = [ join(directory,f) for f in listdir(directory) if isfile(join(directory,f)) ]
	for f in onlyfiles:
		if(f[-3:]=='xml'):
			user_id="user_"+f.split("/")[-1][:-4]
			user_ids.append(user_id)
			print("inserting data for "+f)
			load_gb_file_in_db(f,db_name,xmlschema)
			print("inserted data for "+f+"\n")
	client=MongoClient()
	client[db_name].user_names.insert({"user_ids":user_ids})
	return
```
